#### Is the mean salary per gender different?

#### Find the minimum, mean and the maximum of all numeric columns for each Department.

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
# your answer here
data_employee=employee.groupby('Department',as_index=False).agg({'Salary':['min','mean','max']})
data_employee['difference']=data_employee['Salary']['max']-data_employee['Salary']['min']
data_employee_dif=data_employee[['Department','difference']]
data_employee_dif
```
